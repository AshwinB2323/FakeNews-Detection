```python
import pandas as pd
import os

DATA_PATH = "../data"
```

```python
kaggle_path = os.path.join(DATA_PATH, "FakeNews-Kaggle")
df_fake = pd.read_csv(os.path.join(kaggle_path, "Fake.csv"))
df_real = pd.read_csv(os.path.join(kaggle_path, "True.csv"))

df_fake['label'] = 'fake'
df_real['label'] = 'real'
df_kaggle = pd.concat([df_fake, df_real])

print(f"Kaggle Dataset: {df_kaggle.shape}")
df_kaggle.head()
```

```python
liar_path = os.path.join(DATA_PATH, "LIAR")
df_liar = pd.read_csv(os.path.join(liar_path, "test.tsv"), sep="\t", header=None,
                      names=['id', 'label', 'statement', 'subject', 'speaker', 'job', 'state', 'party',
                             'barely_true', 'false', 'half_true', 'mostly_true', 'pants_on_fire', 'context'])

df_liar = df_liar[['statement', 'label']]
print(f"LIAR Dataset: {df_liar.shape}")
df_liar.head()
```

```python
COAID_PATH = os.path.abspath("../data/CoAID/CoAID-master")
subfolders = [f for f in os.listdir(COAID_PATH) if os.path.isdir(os.path.join(COAID_PATH, f))]

news_fake_list = []
news_real_list = []

for folder in subfolders:
    folder_path = os.path.join(COAID_PATH, folder)
    
    fake_news_path = os.path.join(folder_path, "NewsFakeCOVID-19.csv")
    real_news_path = os.path.join(folder_path, "NewsRealCOVID-19.csv")
    
    if os.path.exists(fake_news_path):
        df_fake = pd.read_csv(fake_news_path)
        df_fake["label"] = "fake"
        news_fake_list.append(df_fake)
    
    if os.path.exists(real_news_path):
        df_real = pd.read_csv(real_news_path)
        df_real["label"] = "real"
        news_real_list.append(df_real)

df_news_fake = pd.concat(news_fake_list, ignore_index=True)
df_news_real = pd.concat(news_real_list, ignore_index=True)

df_coaid_news = pd.concat([df_news_fake, df_news_real], ignore_index=True)

tweets_fake_list = []
tweets_real_list = []

for folder in subfolders:
    folder_path = os.path.join(COAID_PATH, folder)

    fake_tweet_path = os.path.join(folder_path, "NewsFakeCOVID-19_tweets.csv")
    real_tweet_path = os.path.join(folder_path, "NewsRealCOVID-19_tweets.csv")

    if os.path.exists(fake_tweet_path):
        df_fake_tweets = pd.read_csv(fake_tweet_path)
        df_fake_tweets["label"] = "fake"
        tweets_fake_list.append(df_fake_tweets)

    if os.path.exists(real_tweet_path):
        df_real_tweets = pd.read_csv(real_tweet_path)
        df_real_tweets["label"] = "real"
        tweets_real_list.append(df_real_tweets)

df_tweets_fake = pd.concat(tweets_fake_list, ignore_index=True)
df_tweets_real = pd.concat(tweets_real_list, ignore_index=True)

df_coaid_tweets = pd.concat([df_tweets_fake, df_tweets_real], ignore_index=True)
df_coaid_news = df_coaid_news.rename(columns={"title": "headline", "content": "text"})
df_coaid_tweets = df_coaid_tweets.rename(columns={"tweet": "text"})
df_coaid_news = df_coaid_news[['headline', 'text', 'label']]
df_coaid_tweets = df_coaid_tweets[['index', 'label']]

df_coaid_news.head()
```

```python
df_kaggle.to_csv("../data/FakeNews-Kaggle/processed.csv", index=False)
df_liar.to_csv("../data/LIAR/processed.csv", index=False)
df_coaid_news.to_csv("../data/CoAID/news_processed.csv", index=False)
df_coaid_tweets.to_csv("../data/CoAID/tweets_processed.csv", index=False)

print("✅ All datasets saved in their respective folders.")
```
